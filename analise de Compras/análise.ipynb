```python
# ------------------ import base_data -------------------
import pandas

shopping_table = pandas.read_excel("Base Compras.xlsx", sheet_name="fCompras")

display(shopping_table)
```

```python
# -------------------- percentual --------------------------

delivery_date = shopping_table[shopping_table["Previsão Entrega"]>= shopping_table["Data Entrega"]] 

display(delivery_date["Comprador"].value_counts(normalize=True).map("{:.2%}".format))
display(delivery_date)
```

```python
# -------------- gráfico dos compradores que cumpre com o prazo de entrega -------------------
import plotly.express as px

grafico = px.pie(delivery_date, names="Comprador")

grafico.show()
```
